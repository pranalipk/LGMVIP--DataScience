## Data Visualisation

### Terrorist activities by year

```python
count =  df['iyear'].value_counts()
year = count.index.sort_values(ascending  = True)
val = []
for i in year:
    val.append(count[i])
```

```python
plt.figure(figsize = (20,10))
plt.bar(year, val, tick_label = year, color = 'blue')
plt.xticks(rotation  = 90)
plt.title('Number of Terrorists Activities Per Year', fontsize=24)  
plt.xlabel('Year', fontsize = 18, labelpad = 30)
plt.ylabel('Number of terrorists attacks', fontsize = 18, labelpad = 30)
plt.grid(True)
plt.show()
```

### Terrorist Attacks By Region

```python
pd.crosstab(df.iyear, df.region_txt).plot(kind='area',stacked=False,figsize=(15,6))
plt.title('Terrorist Activities By Region In Each Year',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.show()
```

## Number of Attacks on Countries 

```python
plt.subplots(figsize=(8,10))
sns.barplot(df['country_txt'].value_counts()[:16].values,df['country_txt'].value_counts()[:16].index,palette='Paired')
plt.title('Number of Attacks on Countries')
plt.ylabel('Countries')
plt.xlabel('range')
plt.xticks(rotation= 0)
plt.show()
```

### Types of Attacks

```python
attack_type = df.attacktype1_txt.value_counts()[:15]
attack_type
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(attack_type.index,attack_type.values,palette="rocket")
plt.title('Attack Types')
plt.xlabel('Types ot attacks')
plt.ylabel('Number of attacks')
plt.xticks(rotation= 90)
plt.show()
```
